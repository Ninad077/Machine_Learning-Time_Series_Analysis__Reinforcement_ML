# Time Series Analysis (Reinforcement Machine Learning)

**<font color = blue>1. When to use Time Series Analysis?</font>**

Time Series Analysis is used for continuous data and for the data which is auto-correlated.
Autocorrelation means the value in a single column is related to it's previous value.
It is generally used for a data which appears to have a date column (continuous value) & y (dependent col) 
which is autocorrelated.
Having said that Time Series Analysis has only y(dependent variable) & no x (independent variables).
So using this algorithm we will forecast the values of y by analysing the patterns in the data.

**<font color = blue>2. Types of Time Series Analysis?</font>**

Time Series Analysis can be further categorized into 2 types:

    1. Exponential Smoothening
        a. Simple exponential smoothening
        b. Double exponential smoothening
        c. Triple exponential smoothening
        
    2. ARIMA (Autoregressive Integrated Moving Average.)

**<font color = blue>3. Requirements for Time Series Analysis</font>**

The major requirement for the Time series analysis is to first make sure that the data has a unified gaping.
It should be either in daily, weekly or monthly format for the whole data. To give an example, Time series cannot be applied if we have weekly and monthly data together. 

**<font color = blue>4. How should the Time series data look like before preprocessing?</font>**

In Time series analysis we have a Date column which we consider as an index column, in addition to it we have one more
column 'y' (e.g. Sales, Profit,...). 
In a scenario, where the data given to us has multiple columns we have to make sure we only keep the y and date column, 
dropping all other columns.

In Time series analysis we have two important concepts 'Seasonality' and 'Trend'.
Trend for any time series data can be either upward or downward, whereas Seasonality will tell the behaviour of data for 
a particular season (For e.g. Wedding season, Monsoon season, Festivals and so on).
To analyse these patterns, an algorithm called Exponential Smoothing has been developed. 
Exponential Smoothing is categorised into 3 types:

a. Simple Exponential Smoothing

b. Double Exponential Smoothing

c. Triple Exponential Smoothing

Let us understand each smoothening type one by one.

## **<font color = red>11. Exponential Smoothing</font>**

Exponential Smoothing is based on 3 hyperparameters: Alpha (smoothing level:importance to Old values), Beta (smoothing slope:Trend) & Gamma (seasonal period: Seasonality).. Let us try to understand how these parameters influence the smoothing.

### **<font color = green>11.1 Single Exponential Smoothing</font>**

If we have 3 years of data and say if we need to forecast the future 'y' for this data, then it is but natural that the 
recent values from the past will have more importance in forecasting the future values rather than the values that are 
1/2/3 years old.Â 

Alpha is such a hyperparameter which decides how much importance we are giving to the old values. If alpha is close to 0,
then the new values are given more importance and if it is close to 1, then the old values are given more importance.

**Single Exponential Smoothing is the algorithm that deals with only alpha.**
In Single Exponential Smoothing we do not look for Trend (beta) or Seasonality (gamma).

```python
from statsmodels.tsa.api import SimpleExpSmoothing
Exp_Smooth = test.copy()

# smoothing level = alpha
# alpha value should be in the range of 0 to 1
# values close to 0 indicate that older values are very less compared to the recent values.
# values close to 1 indicate that older values are equally important.

fit1 = SimpleExpSmoothing(train).fit(smoothing_level = 0.1)
Exp_Smooth['SES'] = fit1.forecast(steps=len(test))

train.plot(figsize=(10,6),title='Average Sales',fontsize = 14)
test.plot(figsize=(10,6),title='Average Sales',fontsize = 14)
Exp_Smooth['SES'].plot(figsize=(10,6),title='Average Sales',fontsize = 14)
plt.show()
```

Do not forget to first install statsmodels library in the anaconda prompt usingÂ 
pip install statsmodels.
To explain the code above, here we have imported SimpleExpSmoothing function from stats model library.
We have named a variableÂ 'Exp_Smooth'  and saved a copy of test data in it.Â 
Furthermore, we train the model for alpha (smoothing level = 0.1) and fit it later on test data. We use 
Exp_Smooth['SES'] to forecast, here SES stands for Simple Exponential Smoothing.
Lastly we plot the graphs for test, train and Exp_Smooth['SES'].

The line Exp_Smooth['SES'] (green colour) is a line that just forecasts the future sales.
This line however is a straight line and not completely accurate since we have not added the hyperparameter of 
Beta(smoothing slope: trend) & Gamma (seasonal period:Seasonality).


```python
# Checking RMSE for Single Exponential model.

from sklearn.metrics import mean_squared_error
from math import sqrt
rmse = sqrt(mean_squared_error(test,Exp_Smooth.SES))
print(rmse)
```

### **<font color = green>11.2 Double Exponential Smoothing (Holt model)</font>**

**In Double Exponential Smoothing along with alpha (smoothing level) we use the hyperparameter beta (smoothing slope) as well.** 
The only difference in the code would be the smoothing level = 0.25 and fitting the model on the function Holt.
We are further saving the forecast in Holt_df['Holt_linear'] and plotting it with the train & test data

```python
# Holt model
from statsmodels.tsa.api import Holt
Holt_df = test.copy()

# smoothing level = beta
# aplha value should be in the range of 0 to 1
# values close to 0 indicate that older values are very less compared to the recent values.
# values close to 1 indicate that older values are equally important.

fit1 = Holt(train).fit(smoothing_level = 0.1,smoothing_slope = 0.25)
Holt_df['Holt_linear'] = fit1.forecast(steps=len(test))

train.plot(figsize=(10,6),title='Average Sales',fontsize = 14)
test.plot(figsize=(10,6),title='Average Sales',fontsize = 14)
Holt_df['Holt_linear'].plot(figsize=(10,6),title='Average Sales',fontsize = 14)
plt.show()
```

```python
# Checking RMSE for Holt model

from sklearn.metrics import mean_squared_error
from math import sqrt
rmse = sqrt(mean_squared_error(test,Holt_df['Holt_linear']))
print(rmse)
```

### **<font color = green>11.3 Triple Exponential Smoothing (Holt Winter model)</font>**

This is a bit more advanced and efficient algorithm since it performs exponential smoothing on all the 3 hyperparameters. 
**In addition to smoothing line and smoothing slope we introduce another hyperparameter called seasonal period.** 
A seasonal period= 12 indicates 12 months.The code will be the same again, this time however instead of specifying the 
values we are adding the trend and we have defined seasonal periods as 12. The data will be fitted on train model and 
forecasted using the fit on test and stored as 'Holt_Winter_df'.

```python
# Holt Winter model
from statsmodels.tsa.api import ExponentialSmoothing
Holt_Winter_df = test.copy()

# smoothing level = beta
# aplha value should be in the range of 0 to 1
# values close to 0 indicate that older values are very less compared to the recent values.
# values close to 1 indicate that older values are equally important.

fit1 = ExponentialSmoothing(train,seasonal_periods = 12, trend = 'add', seasonal = 'add').fit()
Holt_Winter_df['Holt_linear'] = fit1.forecast(steps=len(test))

train.plot(figsize=(10,6),title='Average Sales',fontsize = 14)
test.plot(figsize=(10,6),title='Average Sales',fontsize = 14)
Holt_Winter_df['Holt_linear'].plot(figsize=(10,6),title='Average Sales',fontsize = 14)
plt.show()
```

If one observes the green graph (Holt_Winter_df['Holt_linear']), one can infer that Triple Exponential Smoothing seems to 
be by far the best smoothing algorithm that forecasts the data more accurately. This can be also confirmed by comparing the 
RMSE value of each Smoothing algorithm.

```python
# Checking RMSE for Holt Winter model.

from sklearn.metrics import mean_squared_error
from math import sqrt
rmse = sqrt(mean_squared_error(test,Holt_Winter_df['Holt_linear']))
print(rmse)
```

## **<font color = red>12. ARIMA (Autoregressive Integrated Moving Average)</font>**

### **<font color = green>12.1 ARIMA concept</font>**

Exponential Smoothening generally works where the **volatility or fluctuations** in the data is not much. However, if we have
volatile data **(say for example Stock market data, cryptocurreny data)** then simple exponential smoothening won't perform 
well. In such scenarios we use ARIMA (Autoregressive Integrated Moving Average).


An ARIMA model is a class of statistical models for analyzing and forecasting time series 
data.

â€¢ **AR:** Autoregression. A model that uses the dependent relationship between an 
          observation and some number of lagged observations.
          
â€¢ **I:** Integrated. The use of differencing of raw observations (e.g. subtracting an 
         observation from an observation at the previous time step) in order to make the
         time series stationary.
         
â€¢ **MA:** Moving Average. A model that uses the dependency between an observation 
          and a residual error from a moving average model applied to lagged observations.
          

### **<font color = green>12.2 Parameters in ARIMA </font>**

The parameters of the ARIMA model are defined as follows:

â€¢ **p:** The number of lag observations included in the model, also called the lag order.

â€¢ **d:** The number of times that the raw observations are differenced, also called the 
         degree of differencing.
         
â€¢ **q:** The size of the moving average window, also called the order of moving average.
         If a Time Series has seasonal patterns, we have to insert seasonal periods, and it 
         becomes SARIMA, short for 'Seasonal ARIMA'.
 
Just like Exponential Smoothening where we had (alpha, beta & gamma), 
here as well by adjusting 'p', 'd' , 'q' values we can adjust the accuracy of the model.

### **<font color = green>12.3 Importing the library pmdarima </font>**

```python
import pmdarima as pm
```

### **<font color = green>12.4 Applying Auto_arima on y & fetching p,d, & q values</font>**

The ARIMA model is an Autoregressive model which means it will optimise the values of 'p', 'd' & 'q' after analysing the
data to give us an accurate forecast.

```python
model = pm.auto_arima(y, start_p = 0,max_p= 3, d=None, max_d = 2, start_q=0, max_q=3, 
                      start_P =0, max_P = 3, D= None, max_D = 2, start_Q = 0, max_Q = 3,
                       max_order= 10, m= 12, seasonal = True, information_criterion = 'aic',
                        test='adf', trace = True, random_state = 10 )
```

In the code block above I am applying auto arima on 'y' data (It is the same data we splitted during Exp Smoothening).
Here, we have to keep the code default without changing any parameters. 
The output of this code will get us the p,d,q values.
In the output 'Best model : ARIMA (0,1,1)', p=0, d = 1, q=1 are the values obtained.

### **<font color = green>12.5 Fitting the model on SARIMAX. </font>**

SARIMAX is a function which is used when we want to fit seasonal data. Since the data is seasonal data related to Sales
we are importing 'SARIMAX' function on which we go ahead and fit the model. 
We have already confirmed the presence of Seasonality from the Holt_Winter model.
Important thing to note here is that we are using'p','d' & 'q' values obtained from auto_arima to fit the model. 
To get started, let us intsall the required statsmodels library package.

```python
! pip install statsmodels
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(train,order=(0,1,1),seasonal_order=(2,0,0,12)).fit()
```

In the code above, I am applying SARIMAX function from statsmodels library and fitting the data on the
'p','d','q' values obtained above where p=0, d=1, q=1. Since we imported SARIMAX we have to specify
(2,0,0) which we obtained after auto_arima followed by seasonal periods = 12 and then fit the model.

As mentioned earlier, auto_arima will itself decide the p,d, q values which then help us in fitting the model as shown above.

### **<font color = green>12.6 Predicting the model using forecast function. </font>**

```python
#pred = model.get_prediction(start-pd.to_datetime('2017-01-31'))

pred = model.get_forecast(steps=12)
ax = y['2014':].plot(label="observed")
pred.predicted_mean.plot(ax=ax, label= "Validation Forecast")
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```

In the code above we are generating a forecast using the get_forecast() method on the fitted model. 
The steps=12 argument indicates that the forecast should be for 12 time steps into the future.
We are then storing this forecast in a variable named 'pred'.

Once stored then we are using data from 'y' to make a plot of observed data. 
The ['2014':] slicing notation is used to select data from the year 2014 onwards. 
The label="observed" in the argument specifies the label for this data series in the plot.

pred.predicted_mean.plot() plots the forecasted values stored in pred on the same plot as the observed data (ax). 
pred.predicted_mean retrieves the predicted mean values from the forecast object pred, and .plot(ax=ax) plots these values
on the same plot as the observed data. 
The label="Validation Forecast" argument specifies the label for predicted mean values of the forecast.

**In simple words we are comparing 2 plots: one is the plot of 'y' 2014 onwards (named as ax:observed) & the second is 
    the predicted mean plot of the forecasted values stored in pred for ax (named as pred.predicted_mean(ax): 
    Validation Forecast)**
    
**Always remember, predicted mean of forecasted values will give us the 'Forecast'. So the above graph in simple words
is a graph comparing actual vs forecasted values.**

Let us see what were the predicted mean values for the forecast.**This predicted mean is nothing but our 'Y_pred'**.

```python
pred.predicted_mean
```

### **<font color = green>12.7 Evaluation of the ARIMA model using rmse. </font>**

Now we are defining our Y_pred & Y_test first, where Y_pred as mentioned is the predicted mean of the forecasted values.
Y_test is the value of 'y' for 2017-01-31 onwards. 
We are then evaluating rmse (root mean squared error) on Y_test & Y_pred.
RMSE will give us the actual difference between the Test & Forecasted values.

```python
Y_pred = pred.predicted_mean
Y_test = y['2017-01-31':]

from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(Y_test,Y_pred))
print(rmse)
```

### **<font color = green>12.8 Evaluation of the ARIMA model using rmse on the entire 'y' data. </font>**

The evaluation we did above was on the model that was fitted on the train data of 'y'. Now let us the ARIMA model for the
entire 'y' data and compare the RMSE values.
To do this, follow the same steps starting from:

1. Fitting the model on *y* on SARIMAX on the 'p','d' & 'q' values obtained after auto_arima. Last time
   we have fitted the model on the **train data of y**.
       
2. Foreacsting the value for the entire data 'y' 2014 onwards. Get the plot of observed vs forecasted values.
    
3. Evaluate the model on Y_test & Y-pred using rmse.

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(y,order=(0,1,1),seasonal_order=(2,0,0,12)).fit()
```

```python
#pred = model.get_prediction(start-pd.to_datetime('2017-01-31'))

pred = model.get_forecast(steps=12)
ax = y['2014':].plot(label="observed")
pred.predicted_mean.plot(ax=ax, label= "Validation Forecast")
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```

```python
pred.predicted_mean
```

```python
Y_pred = pred.predicted_mean
Y_test = y['2017-01-31':]

from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(Y_test,Y_pred))
print(rmse)
```

As seen from above value, we got a much reduced rmse when forecasted on the entire 'y' data instead of chosing the train data.
A lesser rmse indicates better efficiency of the model in predicting accurate results.
